```python
import matplotlib.pyplot as plt
```

```python
days_time = pd.DataFrame(october.groupby(['요일','variable'])['value'].mean())
days_time = days_time.unstack()

plt.figure(figsize=(17,7))
days_time.loc['월', :].plot(label = 'Mon')
days_time.loc['화', :].plot(label = 'Tue')
days_time.loc['수', :].plot(label = 'Wed')
days_time.loc['목', :].plot(label = 'Thu')
days_time.loc['금', :].plot(label = 'Fri')
days_time.loc['토', :].plot(label = 'Sat')
days_time.loc['일', :].plot(label = 'Sun')

plt.xlabel('time')
plt.legend()
plt.plot()
```

```python
#10월 1,2,3 / 7,8,9,10
holiday = october[(october.일자 == 20181001) | (october.일자 == 20181002) | (october.일자 == 20181003) | (october.일자 == 20181007) |
       (october.일자 == 20181008) | (october.일자 == 20181009) | (october.일자 == 20181010)]
```

```python
holiday_time = pd.DataFrame(holiday.groupby(['일자','variable'])['value'].mean())
holiday_time = holiday_time.unstack()
holiday_time
```

```python
plt.figure(figsize=(17,7))
holiday_time.loc[20181001, :].plot(label = '20181001 Mon')
holiday_time.loc[20181002, :].plot(label = '20181002 Tue')
holiday_time.loc[20181003, :].plot(label = '20181003 Wed')
holiday_time.loc[20181007, :].plot(label = '20181007 Sun')
holiday_time.loc[20181008, :].plot(label = '20181008 Mon')
holiday_time.loc[20181009, :].plot(label = '20181009 Tue')
holiday_time.loc[20181010, :].plot(label = '20181010 WEd')

plt.xlabel('time')
plt.legend()
plt.plot()
```

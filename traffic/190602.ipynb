## 8. 미세먼지 구 기준으로 합치기

```python
air2 = pd.read_csv('air02.csv')
air5 = pd.read_csv('air05.csv')
air8 = pd.read_csv('air08.csv')
air11 = pd.read_csv('air11.csv')
```

```python
air2.head(2)
```

```python
air2['date'] = air2['날짜'].apply(lambda x: int(x[:10].replace("-","")))
air2['time'] = air2['날짜'].apply(lambda x: int(x[-2:]))
air5['date'] = air5['날짜'].apply(lambda x: int(x[:10].replace("-","")))
air5['time'] = air5['날짜'].apply(lambda x: int(x[-2:]))
air8['date'] = air8['날짜'].apply(lambda x: int(x[:10].replace("-","")))
air8['time'] = air8['날짜'].apply(lambda x: int(x[-2:]))
air11['date'] = air11['날짜'].apply(lambda x: int(x[:10].replace("-","")))
air11['time'] = air11['날짜'].apply(lambda x: int(x[-2:]))
```

```python
air2.drop(['날짜','시도','측정소코드'],axis=1,inplace=True)
air5.drop(['날짜','시도','측정소코드'],axis=1,inplace=True)
air8.drop(['날짜','시도','측정소코드'],axis=1,inplace=True)
air11.drop(['날짜','시도','측정소코드'],axis=1,inplace=True)
air2.columns = ['district','SO2','CO','O3','NO2','PM10','PM2.5','date','time']
air5.columns = ['district','SO2','CO','O3','NO2','PM10','PM2.5','date','time']
air8.columns = ['district','SO2','CO','O3','NO2','PM10','PM2.5','date','time']
air11.columns = ['district','SO2','CO','O3','NO2','PM10','PM2.5','date','time']
```

```python
def NA_mean(air, col, date, time) :
    return air[(air['date'] == date) & (air['time'] == time)][col].mean()
```

```python
cols = ['SO2','CO','O3','NO2','PM10','PM2.5']
for col in cols :
    frame = air2[air2[col].isnull()]
    mean_value = frame.apply(lambda x : NA_mean(air2, col, x['date'],x['time']), axis =1)
    air2.loc[mean_value.index, col] = mean_value
```

```python
air2.isnull().sum()
```

```python
for col in cols :
    frame = air5[air5[col].isnull()]
    mean_value = frame.apply(lambda x : NA_mean(air5, col, x['date'],x['time']), axis =1)
    air5.loc[mean_value.index, col] = mean_value
```

```python
for col in cols :
    frame = air8[air8[col].isnull()]
    mean_value = frame.apply(lambda x : NA_mean(air8, col, x['date'],x['time']), axis =1)
    air8.loc[mean_value.index, col] = mean_value
```

```python
for col in cols :
    frame = air11[air11[col].isnull()]
    mean_value = frame.apply(lambda x : NA_mean(air11, col, x['date'],x['time']), axis =1)
    air11.loc[mean_value.index, col] = mean_value
```

```python
air11.isnull().sum()
```

```python
car2 = pd.merge(car2,air2,on=['date','district','time'])
car5 = pd.merge(car5,air5,on=['date','district','time'])
car8 = pd.merge(car8,air8,on=['date','district','time'])
car11 = pd.merge(car11,air11,on=['date','district','time'])
```

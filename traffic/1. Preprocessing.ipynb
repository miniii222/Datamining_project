```python
import pandas as pd
car = pd.read_excel('2018년 04월 서울시 차량통행속도.xlsx')
car.drop(['시점명','종점명'],axis=1,inplace=True)
car = pd.melt(car, id_vars=car.columns[:10])
print(car.shape)
car.head()
```

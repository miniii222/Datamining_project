```python
tree = ET.ElementTree(file=urllib.request.urlopen(api_url))
root = tree.getroot()
```

```python
from bs4 import BeautifulSoup
res = urllib.request.urlopen(api_url)
soup = BeautifulSoup(res,"html.parser")
```

```python
print(soup.prettify())
```

```python
variables = ["airline","airportcode","estimateddatetime","flightid","gatenumber","himidity","maxtem","mintem","remart",
             "scheduledatetime","terminalid","wind","yoil"]
```

```python
import pandas as pd
from pandas import DataFrame
airport = pd.DataFrame()
```

```python
airport = airport.append({variables[0]:"","airport":"","airportcode":"",
                         "estimateddatetime":"","scheduledatetime":"","remark":""},ignore_index=True)
```

```python
num = len(soup.find_all("airline"))
for i in range(0,num) :
    airport.set_value(i,"airline",soup.find_all("airline")[i].text)
    airport.set_value(i,"airport",soup.find_all("airport")[i].text)
    airport.set_value(i,"remark",soup.find_all("remark")[i].text)
    airport.set_value(i,"airportCode",soup.find_all("airportcode")[i].text)
        
```

```python
num = len(soup.find_all("airline"))
for name in variables :
    for i in range(0,num) :
        airport.set_value(i,name,soup.find_all(name)[i].text)
```
